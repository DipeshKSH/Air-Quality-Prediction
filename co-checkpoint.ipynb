```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('AirQualityUCI.csv',index_col='Date',na_values=-200)
```

```python
df.info()
```

```python
del df['AH']
del df['Time']
```

```python
df.shape
```

```python
df.describe()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df['CO(GT)']=df['CO(GT)'].fillna(df['CO(GT)'].mean())
df['NOx(GT)']=df['NOx(GT)'].fillna(df['NOx(GT)'].mean())
df['NO2(GT)']=df['NO2(GT)'].fillna(df['NO2(GT)'].mean())
df['PT08.S1(CO)']=df['PT08.S1(CO)'].fillna(df['PT08.S1(CO)'].mean())
df['C6H6(GT)']=df['C6H6(GT)'].fillna(df['C6H6(GT)'].mean())
df['PT08.S2(NMHC)']=df['PT08.S2(NMHC)'].fillna(df['PT08.S2(NMHC)'].mean())
df['PT08.S3(NOx)']=df['PT08.S3(NOx)'].fillna(df['PT08.S3(NOx)'].mean())
df['PT08.S4(NO2)']=df['PT08.S4(NO2)'].fillna(df['PT08.S4(NO2)'].mean())
df['PT08.S5(O3)']=df['PT08.S5(O3)'].fillna(df['PT08.S5(O3)'].mean())
df['T']=df['T'].fillna(df['T'].mean())
df['RH']=df['RH'].fillna(df['RH'].mean())
```

```python
df.isnull().sum()
```

```python
df.drop('NMHC(GT)',axis=1,inplace=True)
```

```python
x = df[['C6H6(GT)','PT08.S2(NMHC)']]
y = df['CO(GT)']
train_x,test_x,train_y,test_y = train_test_split(x,y,test_size=0.2,random_state=90)
```

```python
model = LinearRegression()
```

```python
model.fit(train_x,train_y)
```

```python
model.score(train_x,train_y)
```

```python
model.coef_[0]
```

```python
model.intercept_
```

```python
predict_y=model.predict(test_x)  #the 
predict_y
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(20,10))
ax1 = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122)
#C6H6(GT)','PT08.S2(NMHC)','NOx(GT)','NO2(GT)
ax1.scatter(test_x['C6H6(GT)'], test_x['PT08.S2(NMHC)'], test_y, c='b', marker='o')

ax1.scatter(test_x['C6H6(GT)'], test_x['PT08.S2(NMHC)'], predict_y, c='r',  marker='o')

ax1.set_xlabel('C6H6(GT)')
ax1.set_ylabel('PT08.S2(NMHC)')
ax1.set_zlabel('CO(GT)')


ax2.scatter(test_y,predict_y , color = '#FF1493')
ax2.set_xlabel('Test Y')
ax2.set_ylabel('Predicted Y')
plt.show()
```

```python
mean_squared_error(test_y,predict_y)
```
